*   object : objek apapun yg mengekspos antarmuka array mengembalikan array, atau apa pun (bersarang) urutan

* subok : Secara default, array yg di kembalikan di paksa menjadi array kelas dasar. Jika benar, sub-kelas akan di berikan
* ndimin : Menentukan dimensi minimum dari array yg dihasilkan

membuat array dari yg sudah ada

```python
# konsversi list menjadi ndarray

x = [1, 2, 3, 4, 5]
print(type(x))
print("===konversi list menjadi array===")
x = np.asarray(x)
print(x)
print(x.dtype)
print(type(x))
```

```python
# contoh kedua konversi list menjadi array dengan type data yg di definisikan

x = [1, 2, 3, 4, 5]
print(type(x))

print('konversi list menjadi array dengan type data float')
x = np.asarray(x, dtype='f')
print(x.dtype)
print(x)
```

*   masukkan dua variabel tuple tersebut kedalam variabel bertipe list
* konverse list tersebut menjadi array numpy



```python
# isi di sini
a = (1, 2, 3, 4, 5)
print(type(a))
b = (6, 7, 8, 9, 10)
print(type(b))
print('konversi tuple ke list')
c = list(a) + list(b)
print(type(c))
print(c)
print('konversi list ke array')
c = np.asarray(c)
print(c)
print(type(c))
```
